```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('RnR_GB_DigData_prepared_filtered_IC.csv')
df.head(10)
# df.dropna()
print(df['product_brand'].value_counts())
df[df['product_brand'] == "Ben & Jerry's"].head(10)
```

```python
fig,ax = plt.subplots()
prod = df['product_category']
review = df['review_rating']
benJ = df[df['product_brand'] == "Ben & Jerry's"]


products = benJ['product_name'].to_numpy()
ratings = benJ['review_rating'].to_numpy()

product_sorted=[]
review_score =[]
num_prod = 1
num_review = 1
previous = products[0]
shorthand_values = []
shorthand_index = 0
for i in range(1,len(products)):
    current = products[i]
    if (previous == current):
        #add values to arrays. 
        num_prod = num_prod +1
        num_review = num_review + ratings[i]
    else:
        #save results to arrays
        if(num_prod * num_review >1000):
            #remove all values below 100, a sign they are not popular enough
            product_sorted.append(current[13:])
            review_score.append(num_prod*num_review)
            shorthand_values.append(chr(65+shorthand_index))
            print(shorthand_values[shorthand_index]+": "+ str( review_score[shorthand_index]) +'\n'+ current)
            shorthand_index = shorthand_index +1
            
        num_prod = 1
        num_review = 1
        previous = current
# print(previous)
ax.scatter(shorthand_values,review_score)
y1line = np.ones(len(shorthand_values))
y3000line = y1line * 3000 
ax.set_title('B&J product popularity')
ax.plot(shorthand_values,y3000line,color = 'orange',label = 'y=3000')
ax.set_xlabel('products')
ax.set_ylabel('popularity')
ax.legend()
```

```python
for i in range(len(product_sorted)):
    if ('cookie' in product_sorted[i]):
        print(product_sorted[i] + ': '+str(review_score[i]) +'\n')
```

```python
df = pd.read_csv("brandwatch.csv")
datetime = df['Date']
sentiment = df['Sentiment']
sentiment_processed = []
for i in sentiment:
    if (i == 'negative'):
        sentiment_processed.append(-1)
    if(i == 'neutral'):
        sentiment_processed.append(0)
    if(i=='positive'):
        sentiment_processed.append(1)
        
df.head()
```

```python
import string
from nltk.corpus import stopwords

def remove_punch(mess):
     # Check characters to see if they are in punctuation
    nopunc = [char for char in mess if char not in string.punctuation]

    # Join the characters again to form the string.
    nopunc = ''.join(nopunc)
    
    # Now just remove any stopwords
    return [word for word in nopunc.split() if word.lower() not in stopwords.words('english')]
df['Snippet'].head(5).apply(remove_punch)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
bow_transformer = CountVectorizer(analyzer=remove_punch).fit(df['Snippet'])

# Print total number of vocab words
print(len(bow_transformer.vocabulary_))
```

```python
messages_bow = bow_transformer.transform(df['Snippet'])
```

```python
from sklearn.feature_extraction.text import TfidfTransformer

tfidf_transformer = TfidfTransformer().fit(messages_bow)
```

```python
messages_tfidf = tfidf_transformer.transform(messages_bow)
```

```python
#predicting next value
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(messages_tfidf,sentiment_processed, test_size = 0.33)
clf = MultinomialNB().fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = clf.predict(x_test)
accuracy_score(y_test, y_pred)
```
